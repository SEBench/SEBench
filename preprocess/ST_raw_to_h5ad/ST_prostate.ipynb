```python
import scanpy as sc
import pandas as pd
import anndata as ad
```

```python
# data path of the ST data
data_path = './data/prostate-twelve'
```

```python
# file name of the count matrix and 
count_file = 'P1.1.tsv'
```

```python
# read tsv using pandas
count = pd.read_csv(f'{data_path}/{count_file}', sep='\t')
```

```python
count = count.T
count
```

```python
# get cell_ID
cell_id_array = count.index
cell_id_array
```

```python
# get geneID
gene_id_array = count.columns
```

```python
gene_id_array
```

```python
# get count_matrix
count_X = count.values[:,:].astype('int')
```

```python
# adata
adata = ad.AnnData(count_X)
adata.var_names = gene_id_array
adata.obs_names = cell_id_array
```

```python
adata
```

```python
# get spot coordinate
count['coord_ID'] = count.index
count['x'] = count['coord_ID'].apply(lambda x: x.split('x')[0])
count['y'] = count['coord_ID'].apply(lambda x: x.split('x')[1])
count
```

```python
adata.obsm['spatial'] = count.loc[adata.obs_names].loc[:,['x','y']]
adata
```

```python
adata.write_h5ad(f'{data_path}/prostate_P4.3.h5ad')
```
