```python
import scanpy as sc
import pandas as pd
import anndata as ad
```

```python
data_path = '/home/zhaofy/SEBENCH/Slide-seq/SCP354/other/kidney'
```

```python
count_file = 'dge_hvgs.csv'
meta_file = 'cell_annotations.csv'
```

```python
# 用pandas读取csv
count = pd.read_csv(f'{data_path}/{count_file}')
```

```python
count
```

```python
# 提取cellID
cell_id_array = count['Unnamed: 0']
# 提取geneID
gene_id_array = count.columns
```

```python
gene_id_array
```

```python
# 去掉第一个元素“Unnamed: 0”
gene_id_array = gene_id_array[1:]
gene_id_array
```

```python
# 提取count_matrix
count_X = count.values[:,1:].astype('int')
```

```python
# 构建adata
adata = ad.AnnData(count_X)
adata.var_names = gene_id_array
adata.obs_names = cell_id_array
```

```python
meta = pd.read_csv(f'{data_path}/{meta_file}')
meta
```

```python
# 需要把细胞名作为index，与adata的细胞名align起来
meta = meta.set_index('Unnamed: 0')
```

```python
adata.obs['mouse'] = meta.loc[adata.obs_names].loc[:,'mouse']
adata.obs['type'] = meta.loc[adata.obs_names].loc[:,'type']
adata.obs['id'] = meta.loc[adata.obs_names].loc[:,'id']
```

```python
adata
```

```python
adata.write_h5ad(f'{data_path}/kidney.h5ad')
```
