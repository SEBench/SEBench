```python
data_path1 = 'C:/Users/Administrator/Desktop/code/seqfish+/sourcedata'
data_path2 = 'C:/Users/Administrator/Desktop/code/seqfish+/celltype_annotations'
```

```python
cellcentroids_file = 'cortex_svz_cellcentroids.csv'
counts_file = 'cortex_svz_counts.csv'
annotations_file = 'cortex_svz_cell_type_annotations.csv'
```

```python
cellcentroids = pd.read_csv(f'{data_path1}/{cellcentroids_file}')
counts = pd.read_csv(f'{data_path1}/{counts_file}')
annotations = pd.read_csv(f'{data_path2}/{annotations_file}')
```

```python
id_array = counts._stat_axis
```

```python
gene_id_array = counts.columns
```

```python
gene_id_array
```

```python
count_X = counts.values.astype('float32')
```

```python
adata = ad.AnnData(count_X)
adata.var_names = gene_id_array
adata.obs_names = id_array
```

```python
adata.obs['Field of View'] = cellcentroids.loc[adata.obs_names].loc[:,'Field of View']
adata.obs['Cell ID'] = cellcentroids.loc[adata.obs_names].loc[:,'Cell ID']
adata.obs['Region'] = cellcentroids.loc[adata.obs_names].loc[:,'Region']
adata.obs['louvain'] = annotations.loc[adata.obs_names].loc[:,'louvain']
```

```python
adata.obsm['spatial'] = cellcentroids.loc[adata.obs_names].loc[:,['X','Y']]
```
